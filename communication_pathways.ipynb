```python
import itertools

import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
two_people = nx.Graph()
two_people.add_nodes_from([1, 2])
two_people.add_edge(1,2)

pos = nx.spring_layout(two_people)
```

```python
nx.draw(two_people)
```

```python
ten_nodes = list(range(10))
ten_edges = list(itertools.combinations(ten_nodes, r=2))
```

```python
ten_people = nx.Graph()
ten_people.add_nodes_from(ten_nodes)
ten_people.add_edges_from(ten_edges)

pos=nx.circular_layout(ten_people)
nx.draw(ten_people, pos=pos)
```

```python
def plot_circle_graph(group_size=2):
    group = nx.Graph()
    group.add_nodes_from(range(group_size))
    group.add_edges_from(list(itertools.combinations(range(group_size), r=2)))

    pos=nx.circular_layout(group)
    nx.draw(group, pos=pos)
```

```python
plot_circle_graph(2)
```

```python
plot_circle_graph(5)
```

```python
plot_circle_graph(10)
```
